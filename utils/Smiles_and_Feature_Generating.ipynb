```python
def unique_chemical_extraction (chemicals_excel):
    chemicals = pd.read_excel(f'{chemicals_excel}.xlsx')
    unique_values = chemicals['Compound'].unique()
    # Create a new DataFrame with the unique values
    unique_df = pd.DataFrame({'unique_chemicals': unique_values})

    # Write the new DataFrame to an Excel file
    unique_df.to_excel('unique_chemicals.xlsx', index=False)
```

```python
import requests

def resolve_chemical_name(chemical_name, output_format='smiles'):
    resolver_url = f'https://cactus.nci.nih.gov/chemical/structure/{chemical_name}/{output_format}'
    
    try:
        response = requests.get(resolver_url)
        if response.status_code == 200:
            return response.text
        else:
            return f"Error: Unable to resolve the chemical name. Status code: {response.status_code}"
    except Exception as e:
        return f"Error: {e}"

def chemical_names(chemicals):
#     chemicals = pd.read_excel(f'{chemicals_excel}.xlsx')

    for index,chem in chemicals.iterrows():
        print("========================================")
        print(f"chemical{index}")
#         chemicals.loc[index,'inchi'] = resolve_chemical_name(chem['Compounds'], 'inchi')
#         chemicals.loc[index,'inchikey'] = resolve_chemical_name(chem['Compounds'], 'inchikey')
        chemicals.loc[index,'smiles'] = resolve_chemical_name(chem['Compounds'], 'smiles')
#         chemicals.loc[index,'mol'] = resolve_chemical_name(chem['Compounds'], 'mol')
#         chemicals.loc[index,'cid'] = resolve_chemical_name(chem['Compounds'], 'cid')
#         chemicals.loc[index,'formula'] = resolve_chemical_name(chem['Compounds'], 'formula')
    
    return chemicals
```
